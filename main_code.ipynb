```python
import cv2
face_classifier=cv2.CascadeClassifier(r"C:\Users\hp\Desktop\New folder\ML\part4_facere\haarcascade_frontalface_default.xml")

data=[]
target=[]

for label in labels:
    imgs_path=os.path.join(data_path,label)
    # print(imgs_path)
    img_names=os.listdir(imgs_path)
    # print(img_names)
    # print('---------------------------------')
    for img_name in img_names:
        img_path=os.path.join(imgs_path,img_name)
        print(img_path)
        img=cv2.imread(img_path)
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces=face_classifier.detectMultiScale(gray)

        for(x,y,w,h) in faces:
            # cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
            cropped_face=gray[y:y+h,x:x+w]
            cv2.imshow('cropped_face',cropped_face)
            print('right')
            k=cv2.waitKey(0)
            if(k==121):
                cropped_face=cv2.resize(cropped_face,(50,50))
                data.append(cropped_face)
                target.append(categori_dict[label])
            
        
        # # cv2.imshow('img',img)
        # cv2.imshow('img',cropped_face)
        # # k=cv2.waitKey(100)
        # if(k==27):
        #     break
cv2.destroyAllWindows()
```

```python
data=np.array(data)
data=data.reshape(data.shape[0],data.shape[1]*data.shape[2])
target=np.array(target)

np.save('data',data)
np.save('target',target)

##data tika api hadagattha etika save karanawa
```

```python
import numpy as np
data=np.load(r"C:\Users\hp\Desktop\New folder\ML\part4_facere\data.npy")
target=np.load(r"C:\Users\hp\Desktop\New folder\ML\part4_facere\target.npy")
```

```python
print(data.shape)
print(target.shape)
```

```python
import collections
print(collections.Counter(target))
```

```python
from sklearn.svm import SVC #importing the SVM
from sklearn.decomposition import PCA #importing PCA
from sklearn.pipeline import make_pipeline

pca=PCA(n_components=150,whiten=True,random_state=42)
svc=SVC(kernel='rbf')

model=make_pipeline(pca,svc)
```

```python
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.1)
```

```python
model.fit(train_data,train_target)
```

```python
predicted_target=model.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score

acc=accuracy_score(test_target,predicted_target)
print(acc)
```

```python
from sklearn.metrics import classification_report

print(classification_report(test_target,predicted_target,target_names=['Barack Obama', 'Donald Trump', 'George W Bush']))
```

```python
import joblib

model=joblib.load(r"C:\Users\hp\Desktop\New folder\ML\part4_facere\SVM-Face Recognition my.sav")
```

```python
category_dict={0:'Barack Obama', 1:'Donald Trump',2:'George W Bush'}
```

```python
test_data_path=r"C:\Users\hp\Desktop\New folder\ML\part4_facere\test_data"

import os
import cv2

face_classifier=cv2.CascadeClassifier(r"C:\Users\hp\Desktop\New folder\ML\part4_facere\haarcascade_frontalface_default.xml")

test_img_names=os.listdir(test_data_path)
print(test_img_names)

for test_img in test_img_names:
    img_path=os.path.join(test_data_path,test_img)
    test_img=cv2.imread(img_path)
    
    gray=cv2.cvtColor(test_img,cv2.COLOR_BGR2GRAY)
    
    faces=face_classifier.detectMultiScale(gray)
    
    for (x,y,w,h) in faces:
        
        cv2.rectangle(test_img,(x,y),(x+w,y+h),(0,255,0),2)
        
        cropped_face=gray[y:y+h,x:x+w]
        cropped_face=cv2.resize(cropped_face,(50, 50))
        cropped_original=cropped_face
        cropped_face=cropped_face.reshape(1,50*50)
        
        result=model.predict(cropped_face)
        name=category_dict[result[0]]

        cv2.putText(test_img,name,(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1.5,(0,255,0),2)
        cv2.imshow('GRAY',cropped_original)
    cv2.imshow('LIVE',test_img)
    
    
    k=cv2.waitKey(1000)
    if(k==27):
        break
cv2.destroyAllWindows()
```
